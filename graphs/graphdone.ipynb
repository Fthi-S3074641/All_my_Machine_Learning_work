```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,5,0.1)
y1 = x**2
y2 = x**3
plt.plot(x,y1,color = "blue",linewidth = 5,label = "x^2")
plt.plot(x,y2,color = "red", linewidth = 4,label = "x^3")
plt.xlabel("x")
plt.ylabel("exp(x)")
plt.legend(loc = 2)
plt.text(2,80,"mark",fontsize=10)
#plt.axis([0,100,0,500])
plt.grid()
plt.show()
```

```python
labels = ["A","B","C","D"]
colors = ["Red","Green","brown","Yellow"]
sizes = [4,3,2,1]
explode=[0.04,0.04,0.01,0]
plt.pie(sizes,labels=labels,explode=explode,colors=colors,autopct="%.2f%%",counterclock=False,startangle = 180)
plt.axis("equal")
plt.show()
```
